```python
unmatched_items = []
matched_items = []
for item in progress(item_list):
    l = sku_matcher.match_item(item)
    if not l:
        unmatched_items.append(item)
        continue
    max_score = max([x[1] for x in l])
    matched_items.append((item, list(filter(lambda x: x[1] == max_score, l))))
```
